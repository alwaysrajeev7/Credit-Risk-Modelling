## dealing with df2

```python
index_to_remove = set()

for i in df2.columns.tolist():

    index_to_remove.update(df2[df2[i] == -99999].index.tolist())


len(index_to_remove) # can't remove all of the rows, as major portion of data get vanished
```

```python
# to check which cols has how many -99999 values

count = []

for i in df2.columns.tolist():

    count.append(len(df2[df2[i] == -99999].index))


null_df = pd.DataFrame({'columns':df2.columns.tolist(), '-99999_value':count})
null_df
```

```python
null_df['-99999_value'].values
```

```python
cols_to_remove = null_df[null_df['-99999_value']>10000]['columns'].values.tolist()

print(len(cols_to_remove))

cols_to_remove
```

```python
df2.drop(columns = cols_to_remove, inplace = True) # drop these columns
```

```python
print(df1.shape)
print(df2.shape)
```

```python
# to check which cols has how many -99999 values

count = []

for i in df2.columns.tolist():

    count.append(len(df2[df2[i] == -99999].index))

null_df = pd.DataFrame({'columns':df2.columns.tolist(), 'null_value':count})

null_df

# still this much each feature has -99999 values
```

```python
index_to_remove = set()

for i in df2.columns.tolist():

    index_to_remove.update(df2[df2[i] == -99999].index.tolist())

print(f'No. of rows needed to drop: {len(index_to_remove)}')

print(f'Percentage of rows decied to drop out of total rows: {np.round((len(index_to_remove) / df2.shape[0])*100)} %')
```

```python
df2.drop(index = index_to_remove, inplace = True) # removing these rows
```

```python
print(df1.shape)
print(df2.shape)
```

```python
count = []

for i in df2.columns.tolist():

    count.append(len(df2[df2[i] == -99999].index))

null_df = pd.DataFrame({'columns':df2.columns.tolist(), '-99999_value':count})

null_df # now no more -99999 values
```

# Merging DataFrame

```python
# Merging two DataFrames

df = df1.merge(df2, how = 'inner', on = 'PROSPECTID') # inner join to get common PROSPECTID(customer id) present in both DataFrame

df # merged DataFrame
```

#Checking for duplicate feature

```python
df.T[df.T.duplicated()]
```

```python
duplicate_cols = df.T[df.T.duplicated()].index[0]
duplicate_cols
```

```python
for col in df.columns[1:-1]:

    if np.array_equal(df[col].values, df[duplicate_cols]) and col != duplicate_cols:

        print(f"Column: {duplicate_cols}' is a duplicate of column of: {col}")
```

```python
df[['pct_active_tl','pct_of_active_TLs_ever']]
```

```python
df.drop(columns = duplicate_cols, inplace = True)
```

# Final DataFrame
